# Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in predicting the quality of wine.

The Following are the Key feature ofwine dataset for predicting quality :
    
1- Fixed Acidity: This refers to the concentration of non-volatile acids in the wine. It plays a role in determining the overall tartness and acidity of the wine. Wines with higher fixed acidity might have a more pronounced tart taste.

2- Volatile Acidity: This represents the concentration of volatile acids in the wine. High levels of volatile acidity can result in an unpleasant vinegar-like aroma and taste, negatively affecting wine quality.

3- Citric Acid: Citric acid contributes to the freshness and citrus flavors in wine. It can enhance the overall balance and aroma profile of the wine.

4- Residual Sugar: This is the amount of sugar left in the wine after fermentation. Wines with higher residual sugar content tend to be sweeter, which can affect the perception of balance and mouthfeel.

5- Chlorides: Chlorides contribute to the saltiness of the wine. While a small amount is desirable for enhancing flavors, excessive levels can negatively impact the wine's taste.
    
6- Free Sulfur Dioxide: This compound acts as an antioxidant and preservative. Adequate levels of free sulfur dioxide are important to prevent wine spoilage and maintain freshness.

7- Total Sulfur Dioxide: This combines both free and bound sulfur dioxide. It is another important factor for preserving wine quality and preventing microbial growth.

8- Density: Density is a measure of the wine's mass per unit volume. It can provide insights into the sugar and alcohol content of the wine.

9- pH: pH influences the overall taste, balance, and stability of the wine. Wines with appropriate pH levels tend to have a more pleasant taste and improved aging potential.

10- Sulphates: Sulphates (sulfur compounds) can contribute to the wine's flavor, aroma, and overall quality. They also have an antimicrobial effect, which aids in preserving the wine.

11- Alcohol: Alcohol content affects the wine's body, mouthfeel, and perceived sweetness. It is an essential component in balancing the overall taste.

12- Quality (Target Variable): This is the main variable of interest. It represents the quality rating of the wine, typically on a scale from 0 to 10. This is the variable you're trying to predict based on the other features.
    


# Q2. How did you handle missing data in the wine quality data set during the feature engineering process? Discuss the advantages and disadvantages of different imputation techniques.



1. **Mean/Median/Mode Imputation:**
   - **Advantages:** Simple to implement and can be effective when missing data is minimal. Doesn't distort the distribution of the data significantly.
   - **Disadvantages:** May not be appropriate if the missingness is not random, leading to biased results. Can underestimate the variability if there are substantial missing data.

2. **Forward Fill/Back Fill Imputation:**
   - **Advantages:** Easy to implement. Particularly useful for time-series data where values tend to remain constant between observations.
   - **Disadvantages:** Can introduce temporal order bias if not used carefully. Doesn't work well if missing data patterns are more complex.

3. **Interpolation Techniques (Linear, Polynomial, etc.):**
   - **Advantages:** Preserves relationships between variables. Can provide more accurate estimates if data follows a certain pattern.
   - **Disadvantages:** May not work well if relationships are nonlinear or if the data doesn't follow a clear pattern.

4. **K-Nearest Neighbors (KNN) Imputation:**
   - **Advantages:** Considers relationships between features. Can work well for datasets with complex relationships.
   - **Disadvantages:** Computationally expensive for large datasets. The choice of k (number of neighbors) can impact results. May not work well if the dataset is high-dimensional.

5. **Regression Imputation:**
   - **Advantages:** Uses relationships between variables to estimate missing values. Can be effective when relationships are well-defined.
   - **Disadvantages:** Sensitive to outliers. Requires careful model selection and tuning.

6. **Multiple Imputation:**
   - **Advantages:** Generates multiple imputed datasets to account for uncertainty. Provides a way to propagate imputation uncertainty to downstream analyses.
   - **Disadvantages:** More complex to implement. Requires a good understanding of the dataset and missing data mechanism.

7. **Predictive Modeling Imputation:**
   - **Advantages:** Uses machine learning models to predict missing values. Can capture complex relationships and interactions.
   - **Disadvantages:** Requires significant computational resources. Can introduce model bias if not done carefully.

# Q5. Load the wine quality data set and perform exploratory data analysis (EDA) to identify the distribution of each feature. Which feature(s) exhibit non-normality, and what transformations could be applied to these features to improve normality?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("winequality-red.csv")

w = df.columns.values
number_of_columns = 12
number_of_rows = int(len(w)-1/number_of_columns)

plt.figure(figsize=(2*number_of_columns, 6*number_of_rows))
for i in range(0, len(w)):
    plt.subplot(number_of_rows + 1, number_of_columns, i+1)
    sns.histplot(df[w[i]], kde=True)
```

From the above visualization, the features exhibiting non-normality are:
1- Volatile Acidity

2- citric acid

3- residual sugar

4- chlorides

5- free sulfur dioxide

6- total sulfur dioxide

7- sulphates

8-alcohol


Transformations techniques that can be applied to these features to improve normality are :
1- Log transformation:
    
    - This is one of the most commonly used transformations for normalizing data.
    
    - It is particularly useful when the data is skewed to the right (i.e., positively skewed).

    - A log transformation can help to reduce the skewness of the data by compressing large values and expanding small values.

2- Square root transformation:
    
    - This transformation is useful when the data is skewed to the right and the values are positive.

    - It can help to reduce the skewness and make the data more symmetric.

3- Box-Cox transformation :
    
    - The Box-Cox transformation is useful when the data is skewed and the skewness cannot be corrected by a simple transformation.

    - It involves finding the best transformation parameter lambda that maximizes the normality of the data.

    - This is a family of transformations that includes both the log transformation and the square root transformation.

4- Reciprocal transformation:
    
    - This transformation is useful when the data is skewed to the left (i.e., negatively skewed).

    - It can help to make the data more symmetric.

5- Exponential transformation:
    
    - This transformation is useful when the data is skewed to the left and the values are positive.

    - It can help to make the data more symmetric and reduce the skewness.

# Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("winequality-red.csv")
df.head()
```

```python
df.drop(columns=['quality'], inplace=True)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
s = pd.DataFrame(sc.fit_transform(df), columns=df.columns)
s
```

```python
from sklearn.decomposition import PCA

pca = PCA()
p = pd.DataFrame(pca.fit_transform(s))
p
```

```python
import matplotlib.pyplot as plt
import numpy as np
pd.DataFrame(pca.explained_variance_ratio_).plot.bar()
plt.legend('')
plt.xlabel('Principal Components')
plt.ylabel('Explained Varience');
```

```python

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance');
```

```python
pca_9 = PCA(.9)
pca_9.fit_transform(s)
print("Minimum no of PCA components required to explain ~", round(pca_9.explained_variance_ratio_.sum()*100, 2), "of variance in the data is",pca_9.n_components_, "components.")
```
